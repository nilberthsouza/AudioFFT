```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
def plot_audio_fft(audio_file):
  audio, sr = librosa.load(audio_file)

  fft = librosa.stft(audio)

  log_power = librosa.amplitude_to_db(abs(fft))

  plt.figure(figsize=(10,6))
  librosa.display.specshow(log_power,sr=sr,x_axis='time',y_axis='log')
  plt.title('Espectrograma de Audio')
  plt.xlabel('Tempo')
  plt.ylabel('Frequencia (hz)')
  plt.tight_layout()
  plt.show()


plot_audio_fft('La.mp3')
```

```python
plot_audio_fft('Sol.mp3')
```

```python
plot_audio_fft('acorde.mp3')
```
